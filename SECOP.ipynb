## API Datos SECOP

Este notebook es usado para generar la base de datos los contratos suscritos por entidades departamentales para las vigencias 2016 y 2019. Esto con el fin de determinar el gasto efectivamente ejecutado por cada entidad departamental. Para ello de toma como fuente los datos tomados de la pagina https://www.gov.co/home/

```python
#importación de librerias.

from sodapy import Socrata
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
```

```python
# Solicitamos el contacto con la pagina

client = Socrata("www.datos.gov.co", None)
```

```python
df = client.get("jbjy-vk9h",
                content_type="csv", limit=1000000)
df
```

```python
columns = df[0]
del(df[0])
df = pd.DataFrame(df)
```

```python
df.columns = columns
df
```

```python
# Eliminación de Variables 

BaseSecop= df.drop(['nombre_entidad', 'nit_entidad','ciudad','localizaci_n','rama', 'entidad_centralizada','proceso_de_compra', 'id_contrato', 'referencia_del_contrato',
                   'codigo_de_categoria_principal','descripcion_del_proceso','fecha_de_firma','fecha_de_fin_del_contrato','fecha_de_inicio_de_ejecucion','fecha_de_fin_de_ejecucion', 'condiciones_de_entrega','tipodocproveedor','documento_proveedor','proveedor_adjudicado','es_grupo', 'es_pyme',
                    'habilita_pago_adelantado','liquidaci_n','obligaci_n_ambiental', 'obligaciones_postconsumo', 'reversion','valor_de_pago_adelantado', 'valor_facturado','valor_pendiente_de_pago', 'valor_pagado', 'valor_amortizado',
                    'valor_pendiente_de', 'valor_pendiente_de_ejecucion', 'estado_bpin','c_digo_bpin','saldo_cdp','saldo_vigencia','espostconflicto', 'urlproceso','origen_de_los_recursos', 'dias_adicionados', 'puntos_del_acuerdo',
                    'pilares_del_acuerdo', 'nombre_representante_legal','nacionalidad_representante_legal','tipo_de_identificaci_n_representante_legal','identificaci_n_representante_legal', 'g_nero_representante_legal', 
                    'presupuesto_general_de_la_nacion_pgn','sistema_general_de_participaciones', 'sistema_general_de_regal_as','recursos_propios_alcald_as_gobernaciones_y_resguardos_ind_genas_',
                    'recursos_de_credito', 'recursos_propios', 'ultima_actualizacion','codigo_entidad', 'fecha_inicio_liquidacion', 'fecha_fin_liquidacion','codigo_proveedor'],axis=1)
BaseSecop
```

```python
# Se crea una nueva Base filtrando las Variables orden y destino_gasto

BaseSecop1 = BaseSecop[(BaseSecop.orden=='Territorial')&(BaseSecop.destino_gasto=='Inversión')].copy()
```

```python
#Se revisa el tipo de dato que se tiene en cada variable
BaseSecop1.dtypes
```

```python
#Se cambia el tipo de dato de la variable fecha_de_inicio_del_contrato y la variable Valor de contrato
BaseSecop1['fecha_de_inicio_del_contrato']= pd.to_datetime(BaseSecop1['fecha_de_inicio_del_contrato'])
BaseSecop1['valor_del_contrato']=BaseSecop1['valor_del_contrato'].astype('int')

# Se reemplaza el valor N/D por 0 dentro de la variable anno_bpin
BaseSecop1['anno_bpin'].replace(['N/D'],0,inplace=True)
BaseSecop1['anno_bpin']=BaseSecop1['anno_bpin'].astype('int')
```

```python
# Se remplaza en la variable anno_bpin los valores 'N/D' por 0
BaseSecop1['fecha_de_inicio_del_contrato'].replace(['NaT'],2021,inplace=True)
```

```python
# Se crea la base BaseSecopFinal, con la que se van a realizar las estimaciones pertinentes
BaseSecop2 = BaseSecop1.drop(['orden','sector','estado_contrato','tipo_de_contrato','modalidad_de_contratacion','justificacion_modalidad_de','fecha_de_inicio_del_contrato','destino_gasto'],axis=1)
BaseFinal=BaseSecop2[(BaseSecop2.anno_bpin>1)&(BaseSecop2.anno_bpin<2020)]
```

```python
BaseFinal
```

```python
BaseFinal.describe()
```

```python
#se coloca como indice los departamentos de Colombia
BaseFinal=BaseFinal.groupby(["departamento","anno_bpin"]).sum()
```

```python
#Se descarga la base final con el fin de poderla llamar en el siguiente proyecto.

BaseFinal.to_csv('Base_Secop.csv')
```
